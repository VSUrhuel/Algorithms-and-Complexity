```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tabulate import tabulate
a = []
b = []
c = []
d = []
e = []
f = []
#Reading the Text File
for line in open("insertFIle.txt", "r"):
    lines = [i for i in line.split()]
    a.append((float(lines[0]))/1000)
    b.append(float(lines[1]))
    c.append(float(lines[2]))
    d.append(float(lines[3]))
#Plotting the Run Time
plt.title("Insertion")
plt.xlabel("Size of Input (in thousands)")
plt.ylabel("Time (t)")
plt.plot(a, b, c = 'g', label = 'BST')
plt.plot(a, c, c = 'b', label = 'AVL')
plt.plot(a, d, c = 'r', label = 'RBT')
plt.legend()
plt.show()
```
