```python
import matplotlib.pyplot as plt
import numpy as np
import re

#Reading the File
file = open("test.txt", "r")
file2 = open("theorize.txt", "r")
content2 = file2.read()
content = file.read()

lines2 = content2.splitlines()
lines = content.splitlines()
x = [] 
x2 = []
y = [] 
y2 = []

#Displaying the File Content
print("Actual Runtime")
print("   Input    Time")
for l in lines:
    nums = (re.split(r' - ', l))
    print(nums)
    x.append(int(nums[0]))
    y.append(float(nums[1]))
print("\nAsymptotic Runtime")
print("   Input    Time")
for l2 in lines2:
    nums2 = (re.split(r' - ', l2))
    print(nums2)
    x2.append(int(nums2[0]))
    y2.append(float(nums2[1]))
    
file.close()
```

```python
#Displaying the Run time Algorithm
plt.title ("Running time of algorithm")
plt.xlabel ("Size of Input");
plt.ylabel ("Time")
plt.plot(x, y, 'b-')
plt.plot(x2, y2, 'g-')
plt.legend(["Actual", "Asymptotic"])
plt.plot(x2, y2, 'go')
plt.plot(x, y, 'bo')
plt.show()
```
