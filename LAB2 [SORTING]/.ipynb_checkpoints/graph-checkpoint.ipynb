```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tabulate import tabulate
a = []
b = []
c = []
d = []
e = []
f = []
#Reading the Text File
for line in open("file.txt", "r"):
    lines = [i for i in line.split()]
    a.append((float(lines[0])/1000))
    b.append(float(lines[1]))
    c.append(float(lines[2]))
    d.append(float(lines[3]))
    e.append(float(lines[4]))
    f.append(float(lines[5]))
#Plotting the Run Time
plt.title("Sorting Algroithms")
plt.xlabel("Size of Input (in thousands)")
plt.ylabel("Time (t)")
plt.plot(a, b, marker = 'o', c = 'g', label = 'BubbleSort - O(n^2)')
plt.plot(a, c, marker = 'o', c = 'b', label = 'InsertionSort - O(n^2)')
plt.plot(a, d, marker = 'o', c = 'k', label = 'SelectionSort - O(n^2)')
plt.plot(a, e, marker = 'o', c = 'r', label = 'MergeSort - O(log n)')
plt.plot(a, f, marker = 'o', c = 'y', label = 'QuickSort - O(log n)')
plt.legend()
plt.show()
print("\t\tFigure 1. Run time of Sorting Algorithms")
```

```python
#Display Table
print("  Table 1. Run time of each sorting algorithm in various inputs")
print(tabulate({"Input": a, "BSSort": b, "ISSort": c, "SSSort": d, "MSSort":e, "QSSort":f}, headers = "keys", tablefmt="orgtbl"))
```

```python
#Analysis
print("Shown in the graph is the run time interval of each sorting algorithm.")
print("Important Points to Take Note:")
print("\t*The slowest algorithm among any sorting algorithm is Bubble Sort.")
print("\t*The fastest algorithm among any sorting algorithm is Quick Sort and Merge Sort being closely the same.")
print("\t*Among O(n^2) sorting algorithm, the Bubble Sort is slowest.")
print("\t*Among O(n^2) sorting algorithm, the Insertion Sort is fastest.")
```
