```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tabulate import tabulate
a = []
b = []
c = []
#Reading the Text File
for line in open("file.txt", "r"):
    lines = [i for i in line.split(',')]
    a.append(float(lines[0]))
    b.append(float(lines[1]))
    c.append(float(lines[2]))
#Plotting the Run Time
plt.title("Graph Search Algorithm")
plt.xlabel("Size of Input")
plt.ylabel("Time (t)")
plt.plot(a, b, marker = 'o', c = 'g', label = 'Breadth First Search')
plt.plot(a, c, marker = 'o', c = 'b', label = 'Depth First Search')
plt.legend()
plt.show()
print("\t\tFigure 1. Runtime for Graph Search")
```
