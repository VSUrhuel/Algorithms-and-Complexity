```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tabulate import tabulate
a = []
b = []
c = []
d = []
#Reading the Text File
for line in open("file.txt", "r"):
    lines = [i for i in line.split(',')]
    a.append(float(lines[0]))
    b.append(float(lines[1]))
    c.append(float(lines[2]))
    d.append(float(lines[3]))
#Plotting the Run Time
plt.title("Graphs Algorithm")
plt.xlabel("Size of Input")
plt.ylabel("Time (t)")
plt.plot(a, c, marker = 'o', c = 'b', label = 'Prims')
plt.plot(a, b, marker = 'o', c = 'g', label = 'Kruskal')
plt.plot(a, d, marker = 'o', c = 'r', label = 'Boruvkas')
plt.legend()
plt.show()
print("\t\tFigure 1. Runtime for Graph Search")
```
