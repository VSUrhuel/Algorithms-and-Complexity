# Analysis of Algorithm for Minimum Spanning Trees
- The figure below shows the run time of three different algorithms for finding the minimum spanning trees, namely, Prim's, Kruskal's, and Boruvka's algorithms. The input ranges from 700 to 1700 random edges, with 10 being the maximum weight.
  
## Prim's Algorithm
- It is shown in Figure 1 that Prim's algorithm is the slowest for the spanning trees algorithm, wherein at the maximum input, it reaches 19 seconds.
- One factor that affects the runtime of Prim's algorithm is its worst-case time complexity which is at O(V^2).
- However, in practice, Prim's algorithm can be reduced to O(ElogV).
- Furthermore, Prim's Algorithm also involved more complex implementation, which increases its factors to increase its runtime.

## Kruskal's Algorithm
- Based on the graph, Kruskal's algorithm is not that fast yet slow, wherein at its maximum set of input, it reaches 13 seconds.
- The time complexity of this algorithm is at O(ElogV), which is in logarithm time which is a great factor in its runtime.
- This is in logarithmic time because of its edge sorting, wherein the algorithm sorts the edge weights.
- This is generally simple in terms of its structure implementation among others.

## Boruvka's Algorithm
- Evident in the graph that the fastest algorithm among others is Boruvka's algorithm, wherein the range of its runtime does not go up to two seconds in different sets of inputs.
- A big contributing factor to its runtime is its O(ElogV) runtime in its worst case.
- Furthermore, this offers better potential for parallelization, leading to further performance gains in suitable environments.

## Conclusion
- The choice of algorithm for finding the minimum spanning tree depends on various factors.
- While Boruvka's algorithm shines in terms of speed, Kruskal's simplicity and good performance across different graph densities often make it a popular choice.
- On the other hand, Prim's algorithm might be suitable for specific cases but less efficient.
- Hence, when selecting an algorithm, it is not the runtime that should be considered, it is recommended also to consider the specific characteristics of the graph, implementation details, and potential for parallelization.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tabulate import tabulate
a = []
b = []
c = []
d = []
#Reading the Text File
for line in open("file.txt", "r"):
    lines = [i for i in line.split(',')]
    a.append(float(lines[0]))
    b.append(float(lines[1]))
    c.append(float(lines[2]))
    d.append(float(lines[3]))
#Plotting the Run Time
plt.title("Minimum Spanning Trees Algorithm")
plt.xlabel("Size of Input")
plt.ylabel("Time (t)")
plt.plot(a, c, marker = 'o', c = 'b', label = "Prim's")
plt.plot(a, b, marker = 'o', c = 'g', label = "Kruskal's")
plt.plot(a, d, marker = 'o', c = 'r', label = "Boruvka's")
plt.legend()
plt.show()
print("\t\tFigure 1. Runtime for Minimum Spanning Trees Algorithm")
```
